# 1. Chargement des donnÃ©es

```python
def load_data(image_dir, mask_dir):
    images, masks = [], []
    for file in os.listdir(image_dir):
        img_file_path = os.path.join(image_dir, file)
        mask_file_path = os.path.join(mask_dir, file.replace('images', 'masks'))

        if os.path.exists(mask_file_path):
            image = np.load(img_file_path)
            mask = np.load(mask_file_path)
            images.append(cv2.resize(image, (256, 256)))
            # Redimensionner le masque en utilisant l'interpolation du voisinage le plus proche
            # afin de prÃ©server l'intÃ©gritÃ© des Ã©tiquettes
            masks.append(cv2.resize(mask, (256, 256), interpolation=cv2.INTER_NEAREST))
        else:
            print(f"Skipping {img_file_path} due to missing mask.")

    if len(images) == 0 or len(masks) == 0:
        print(f"Warning: No images or masks found in {image_dir}. Returning empty arrays.")
        return np.empty((0, 256, 256)), np.empty((0, 256, 256))

    return np.array(images), np.array(masks)


```

```python
train_image_dir = "/content/data/train/images"
train_mask_dir = "/content/data/train/labels"
test_image_dir = "/content/data/test/images"
test_mask_dir = "/content/data/test/labels"
val_image_dir = "/content/data/val/images"
val_mask_dir = "/content/data/val/labels"
```

```python
# Charger les donnÃ©es
X_train, y_train = load_data(train_image_dir, train_mask_dir)
X_test, y_test = load_data(test_image_dir, test_mask_dir)
X_val, y_val = load_data(val_image_dir, val_mask_dir)
```

# 1. Fonctions de perte avancÃ©es


**Dice Loss :**
Mesure la similaritÃ© entre les masques rÃ©els et prÃ©dits.

Utile pour les problÃ¨mes de segmentation avec dÃ©sÃ©quilibre des classes.

**Focal Loss :**

Donne plus de poids aux exemples difficiles Ã  classer.

Utile pour les problÃ¨mes de classification avec dÃ©sÃ©quilibre des classes.

```python
def dice_loss(y_true, y_pred, smooth=1e-6):
    # Aplatir les tenseurs y_true et y_pred pour les transformer en vecteurs 1D
    y_true_f = K.flatten(y_true)
    y_pred_f = K.flatten(y_pred)

    # Calculer l'intersection entre les valeurs vraies et prÃ©dites
    intersection = K.sum(y_true_f * y_pred_f)

    # Calculer le coefficient de Dice
    dice = (2. * intersection + smooth) / (K.sum(y_true_f) + K.sum(y_pred_f) + smooth)

    # Retourner la perte de Dice (1 - Dice)
    return 1 - dice

def focal_loss(gamma=2., alpha=0.25):
    def focal_loss_fixed(y_true, y_pred):
        # Obtenir une petite constante epsilon pour Ã©viter la division par zÃ©ro
        epsilon = K.epsilon()

        # Clipper les prÃ©dictions pour Ã©viter les valeurs extrÃªmes
        y_pred = K.clip(y_pred, epsilon, 1. - epsilon)

        # Calculer pt, qui est la probabilitÃ© prÃ©dit pour les classe 1 ou 0
        pt = tf.where(K.equal(y_true, 1), y_pred, 1 - y_pred)

        # Calculer la perte focal
    return focal_loss_fixed
```

# 3. DÃ©finition du modÃ¨le U-Net avec Dice Loss


```python
sfrom tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, UpSampling2D, concatenate, Multiply, Add
from tensorflow.keras.models import Model



def unet_model(input_size=(256, 256, 1)):
    inputs = Input(input_size)
    # Bloc 1 : Convolution et MaxPooling
    c1 = Conv2D(64, (3, 3), activation='relu', padding='same')(inputs)
    c1 = Conv2D(64, (3, 3), activation='relu', padding='same')(c1)
    p1 = MaxPooling2D((2, 2))(c1)
    # Bloc 2 : Convolution et MaxPooling
    c2 = Conv2D(128, (3, 3), activation='relu', padding='same')(p1)
    c2 = Conv2D(128, (3, 3), activation='relu', padding='same')(c2)
    p2 = MaxPooling2D((2, 2))(c2)
    # Bloc 3 : Convolution et MaxPooling
    c3 = Conv2D(256, (3, 3), activation='relu', padding='same')(p2)
    c3 = Conv2D(256, (3, 3), activation='relu', padding='same')(c3)
    p3 = MaxPooling2D((2, 2))(c3)
    # Bloc 4 : Convolution et MaxPooling
    c4 = Conv2D(512, (3, 3), activation='relu', padding='same')(p3)
    c4 = Conv2D(512, (3, 3), activation='relu', padding='same')(c4)
    p4 = MaxPooling2D((2, 2))(c4)
    # Bloc 5 : Convolution au centre du U-Net
    c5 = Conv2D(1024, (3, 3), activation='relu', padding='same')(p4)
    c5 = Conv2D(1024, (3, 3), activation='relu', padding='same')(c5)
    # Bloc de dÃ©codage : UpSampling et concatÃ©nation
    u6 = UpSampling2D((2, 2))(c5)
    u6 = concatenate([u6, c4])
    c6 = Conv2D(512, (3, 3), activation='relu', padding='same')(u6)
    c6 = Conv2D(512, (3, 3), activation='relu', padding='same')(c6)

    u7 = UpSampling2D((2, 2))(c6)
    u7 = concatenate([u7, c3])
    c7 = Conv2D(256, (3, 3), activation='relu', padding='same')(u7)
    c7 = Conv2D(256, (3, 3), activation='relu', padding='same')(c7)

    u8 = UpSampling2D((2, 2))(c7)
    u8 = concatenate([u8, c2])
    c8 = Conv2D(128, (3, 3), activation='relu', padding='same')(u8)
    c8 = Conv2D(128, (3, 3), activation='relu', padding='same')(c8)

    u9 = UpSampling2D((2, 2))(c8)
    u9 = concatenate([u9, c1])
    c9 = Conv2D(64, (3, 3), activation='relu', padding='same')(u9)
    c9 = Conv2D(64, (3, 3), activation='relu', padding='same')(c9)

    outputs = Conv2D(1, (1, 1), activation='sigmoid')(c9)

    model = Model(inputs=[inputs], outputs=[outputs])
    return model


# CrÃ©er le modÃ¨le
model = unet_model(input_size=(256, 256, 1))
model.compile(optimizer='adam', loss=focal_loss(gamma=2.0, alpha=0.25), metrics=['accuracy'])
model.summary()
```

# 5. Callbacks pour un meilleur entraÃ®nement


#**Callbacks**


ArrÃªte l'entraÃ®nement si la perte de validation ne s'amÃ©liore pas pendant 10 Ã©poques.

Restaure les poids du meilleur modÃ¨le.


```python
callbacks = [
    EarlyStopping(patience=10, verbose=1, restore_best_weights=True),
    ReduceLROnPlateau(factor=0.1, patience=5, min_lr=1e-6, verbose=1),
    ModelCheckpoint('best_model.keras', verbose=1, save_best_only=True)
]
```

# 6. EntraÃ®nement du modÃ¨le


```python
history = model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    batch_size=16,
    epochs=50,
    callbacks=callbacks
)
```
